```python
appearences_hash = {}
for index, row in appearences_df.iterrows():
    key = '{}_{}'.format(row.playerID, row.yearID)
    if key not in appearences_hash:
        appearences_hash[key] = 1
    else:
        appearences_hash[key] += 1
```

```python
more_than_once = {}
pid_yearid = []
for key, value in appearences_hash.items():
    if value >= 2:
        more_than_once[key] = value
        pid_yearid.append(key.split('_'))
```

```python
multiple_df = pd.DataFrame(pid_yearid, columns=['playerID', 'yearID'])
#convert to numerics
multiple_df['yearID'] = multiple_df['yearID'].astype(np.int64)
```

```python
batting_df = pd.read_csv('../data/baseballdatabank/core/Batting.csv')
display(batting_df.dtypes)
display(multiple_df.dtypes)
```

```python
multiple_df_batting = multiple_df.merge(right=batting_df, on=['playerID', 'yearID'], how='left')
multiple_df_batting.tail(20)
```
